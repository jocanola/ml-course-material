## Add and remove column


```python
# Column from series
seats_column = pd.Series([5,5,5,5,5,5])

# New colument called seats
car_sales["Seats"] = seats_column
car_sales["Seats"] = car_sales["Seats"].fillna(5)
car_sales
```

```python
# Column from python list
fuel_economy = [4,4,4,5.8,4,4,4.2,4.3,4.4, 4.5]
car_sales["Fuel per 100km"] = fuel_economy
car_sales
```

```python
# Column from another existing column

car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100km"]
car_sales
```
